```python
import cobra
```

```python
from cobra import Model,Reaction,Metabolite
```

```python
# Glycolysis pathway

model=Model('example')
HK=Reaction('HK')
HK.name='Hexokinase'
HK.lower_bound=0
HK.upper_bound=1000


PGI=Reaction('PGI')
PGI.name='Glucose 6-phosphate isomerase'
PGI.lower_bound=-1000
PGI.upper_bound=1000


PFK=Reaction('PFK')
PFK.name='Phosphofructokinase'
PFK.lower_bound=0
PFK.upper_bound=1000


TPI=Reaction('TPI')
TPI.name='Triose-phosphate isomerase'
TPI.lower_bound=-1000
TPI.upper_bound=1000


ALD=Reaction('ALD')
ALD.name='Fructose 1,6-diphosphate aldolase'
ALD.lower_bound=-1000
ALD.upper_bound=1000


GAPDH=Reaction('GAPDH')
GAPDH.name='Glyceraldehyde 3-phosphate dehydrogenase'
GAPDH.lower_bound=-1000
GAPDH.upper_bound=1000


PGK=Reaction('PGK')
PGK.name='Phosphoglycerate kinase'
PGK.lower_bound=-1000
PGK.upper_bound=1000


PGLM=Reaction('PGLM')
PGLM.name='Phosphoglycerate mutase'
PGLM.lower_bound=-1000
PGLM.upper_bound=1000


ENO=Reaction('ENO')
ENO.name='Enolase'
ENO.lower_bound=-1000
ENO.upper_bound=1000


PK=Reaction('PK')
PK.name='Pyruvate kinase'
PK.lower_bound=0
PK.upper_bound=1000



AMP_R=Reaction('AMP')
AMP_R.name='AMP export'
AMP_R.lower_bound=0
AMP_R.upper_bound=1000


APK=Reaction('APK')
APK.name='Adenylate kinase'
APK.lower_bound=-1000
APK.upper_bound=1000


PYR_R=Reaction('PYR_R')
PYR_R.name='Adenylate kinase'
PYR_R.lower_bound=-1000
PYR_R.upper_bound=1000



ATP_R=Reaction('ATP')
ATP_R.name='ATP hydrolysis'
ATP_R.lower_bound=0
ATP_R.upper_bound=1000


NADH_R=Reaction('NADH')
NADH_R.name='ATP hydrolysis'
NADH_R.lower_bound=0
NADH_R.upper_bound=1000


GLUin=Reaction('GLUin')
GLUin.name='Glucose import'
GLUin.lower_bound=1
GLUin.upper_bound=1


AMPin=Reaction('AMPin')
AMPin.name='Glucose import'
AMPin.lower_bound=0
AMPin.upper_bound=1000


H_R=Reaction('H')
H_R.name='Proton exchange'
H_R.lower_bound=-1000
H_R.upper_bound=1000


H2O_R=Reaction('H2O')
H2O_R.name='Water exchange'
H2O_R.lower_bound=-1000
H2O_R.upper_bound=1000


#missed reactions-----
LDH=Reaction('LDH')
LDH.name='Lactate dehydrogenase'
LDH.lower_bound=-1000
LDH.upper_bound=1000


LAC_R=Reaction('LAC_R')
LAC_R.name='Lactate exchange'
LAC_R.lower_bound=-1000
LAC_R.upper_bound=1000
#----------------------
```

```python
####### Metabolite  ###########################


Gluc = Metabolite(
    'Gluc',
    formula='C6H12O6',
    name='Glucose',
    compartment='c')

G6P=Metabolite(
    'G6P',
    formula='C6H13O9P',
    name='Glucose 6-phosphate',
    compartment='c')

ATP=Metabolite(
    'ATP',
    formula='C10H12N5O13P3',
    name='Adenosine tri-phosphate',
    compartment='c')

ADP=Metabolite(
    'ADP',
    formula='C10H15N5O10P2',
    name='Adenosine di-phosphate',
    compartment='c')
H=Metabolite(
    'H',
    formula=' H',
    name='Proton',
    compartment='c')
F6P = Metabolite(
    'F6P',
    formula='C6H13O9P',
    name='Fructose 6-phosphate',
    compartment='c')
FDP = Metabolite(
    'FDP',
    formula='C6H14O12P2',
    name='Fructose 1,6-diphosphate',
    compartment='c')
DHAP = Metabolite(
    'DHAP',
    formula='C3H7O6P',
    name='Dihydroxyacetone phosphate',
    compartment='c')
GAP = Metabolite(
    'GAP',
    formula='C3H7O6P',
    name='Glyceraldehyde 3-phosphate',
    compartment='c')

DPG13 = Metabolite(
    'DPG13',
    formula='C3H8O10P2',
    name='1,3-Diphosphoglycerate',
    compartment='c')
PG3 = Metabolite(
    'PG3',
    formula='C3H7O7P',
    name='3-Phosphoglycerate',
    compartment='c')
PG2 = Metabolite(
    'PG2',
    formula='C3H7O7P',
    name='2-Phosphoglycerate',
    compartment='c')
PEP = Metabolite(
    'PEP',
    formula=' C3H2O6P',
    name='Phosphoenolpyruvate',
    compartment='c')
PYR = Metabolite(
    'PYR',
    formula='C3H3O3',
    name='Pyruvate',
    compartment='c')

NAD = Metabolite(
    'NAD',
    formula='C21H26N7O14P2',
    name='Nicotinamide adenine dinucleotide (oxidized)',
    compartment='c')
NADH = Metabolite(
    'NADH',
    formula='C21H27N7O14P2',
    name='Nicotinamide adenine dinucleotide (reduced)',
    compartment='c')
AMP = Metabolite(
    'AMP',
    formula='C10H12N5O7P',
    name='Adenosine mono-phosphate',
    compartment='c')
P = Metabolite(
    'P',
    formula='PO4',
    name='Inorganic phosphate',
    compartment='c')

H2O= Metabolite(
    'H2O',
    formula='H2O',
    name='Water',
    compartment='c')
#---------------------
LAC = Metabolite(
    'LAC',
    formula='C3H6O3',
    name='Lactate',
    compartment='c')
#---------------------
```

```python
############# add Metabolites To Reactions ##################################3


HK.add_metabolites({
     Gluc:-1.0,
     ATP:-1.0,
     G6P:1.0,
     ADP:1.0,
     H:1.0
})


PGI.add_metabolites({
     G6P:-1.0,
     F6P:1.0,
     
})


PFK.add_metabolites({
    F6P:-1,
    ATP:-1 ,
    FDP :1,
    ADP :1,
    H:1   ,
})

TPI.add_metabolites({
    DHAP:-1,
    GAP:1 ,
})

ALD.add_metabolites({
    FDP:-1,
    DHAP:1 ,
    GAP :1  ,
})

GAPDH.add_metabolites({
    GAP:-1,
    NAD:-1 ,
    P:-1,
    DPG13 :1,
    NADH :1,
    H:1,
})

PGK.add_metabolites({
    DPG13:-1,
    ADP:-1 ,
    PG3 :1,
    ATP :1,
})

PGLM.add_metabolites({
    PG3:-1,
    PG2:1 ,
})

ENO.add_metabolites({
    PG2:-1,
    PEP:1 ,
    H2O:1,
})

PK.add_metabolites({
    PEP:-1,
    ADP:-1, 
    H :-1,
    PYR :1,
    ATP:1  , 
})



AMP_R.add_metabolites({
    AMP:-1,
  
})

APK.add_metabolites({
    ADP:-2,
    AMP:1 ,
    ATP :1,
})

PYR_R.add_metabolites({
    PYR:-1
})


ATP_R.add_metabolites({
    ATP :-1,
    H2O:-1,
    ADP:1,
    P:1,
    H:1,
})

NADH_R.add_metabolites({
    NADH :-1,
    NAD:1,
    H:1,
})

GLUin.add_metabolites({
    Gluc:1, 
})

AMPin.add_metabolites({
    AMP:1 ,
})

H_R.add_metabolites({
    H:-1 ,
})


H2O_R.add_metabolites({
    H2O:-1, 
})

#--------------------
LDH.add_metabolites({
    PYR:-1,
    NADH:-1,
    H:-1,
    LAC:1,
    NAD:1,
})

LAC_R.add_metabolites({
    LAC:-1
})
```

```python
############ Add  Reactions to Model ##########################

model.add_reactions([HK,PGI,PFK,TPI,ALD,GAPDH,PGK,PGLM,ENO,PK,AMP_R,APK,PYR_R,ATP_R,NADH_R,GLUin,AMPin,H_R,H2O_R,
                   LDH, LAC_R])
```

```python
### Add  objective Function #####################

model.objective = 'LAC_R'
```

```python
model.optimize()
```
